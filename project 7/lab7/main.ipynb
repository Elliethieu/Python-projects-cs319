```python
import csv

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

data=process_csv("vaccinations.csv")
```

```python
header=data[0]
header
```

```python
covid_data=data[1:]
```

```python
def cell(row_idx, col_name):
    col_idx = header.index(col_name)
    val = covid_data[row_idx][col_idx]
    if val == "":
        return None
    #if col_idx == csv_header.index("latitude") or col_idx == csv_header.index("longitude") or col_idx == csv_header.index("reviews_per_month"):
    #    return float(val)
    if col_idx == header.index("daily_vaccinations") or col_idx == header.index("total_vaccinations") or col_idx == header.index("people_vaccinated") or col_idx == header.index("people_fully_vaccinated")  or col_idx == header.index("population"):
        return int(val)

    # optional: convert types based on column name, 
    # ensuring that the float and int values are appropriately converted
    return val
```

```python
#Q1: What dates are mentioned in the dataset?
dates=[]

for idx in range(len(covid_data)):
    dates.append(cell(idx, "date"))

dates = sorted(list(set(dates)))

dates   
```

```python
#Q2: What countries are listed in the dataset?
countries=[]

for idx in range(len(covid_data)):
    countries.append(cell(idx, "country"))

countries = sorted(list(set(countries)))

countries  
```

```python
#Q3: What are the populations of all the countries in the dataset?
populations = {} #key: country name, value: popluation

for row_idx in range(len(covid_data)):
    country = cell(row_idx, "country")
    population = cell(row_idx, "population")
    if country not in populations:
        populations[country] = None # Initial place holder for country, we will replace this using next condition
    if population != None: # TODO: check that the population data is not missing
        populations[country] = population # TODO: add the key-value pair to the dict

populations      
```

```python
#Q4: How many countries have daily vaccination date for the date '10/12/2021'?
daily_vaccination_10_12_2021=[]

for idx in range(len(covid_data)):
    if cell(idx,"date") == '10/12/2021':
        if cell(idx,"daily_vaccinations")!= None:
            daily_vaccination_10_12_2021.append(cell(idx,"country"))

daily_vaccination_10_12_2021_2= list(set(daily_vaccination_10_12_2021))

len(daily_vaccination_10_12_2021_2)         
```

```python
def daily_vaccinations_on(given_date):
    '''return a dictionary mapping each country to 
    the number of daily vaccinations on the given date; 
    if data is missing, the value for that country is None'''
    daily_vax_dict = {}
    for i in range(len(covid_data)):
        date = cell(i, "date")
        if date == given_date:
	# if date equals given_date
            country =  cell(i,"country")
            daily_vax =  cell(i,"daily_vaccinations")
            #print(date, country, daily_vax)  # use this for debugging
            if daily_vax == None:
                daily_vax_dict[country]= None
                # if daily_vax is the empty string: 
                # see docstring above
            else:
                daily_vax_dict[country]= daily_vax       
                # convert to an int and add this value to the dictionary
    return daily_vax_dict

```

```python
#Q6: How many vaccinations are estimated to have been done for each country?
total_daily_vaccination={}

for i in range(len(covid_data)):
    if cell(i,"country") not in total_daily_vaccination:
        total_daily_vaccination[cell(i,"country")]= None
    if cell(i, "daily_vaccinations") == None:
        continue
    else:
        daily_add = cell(i, "daily_vaccinations") 
        if total_daily_vaccination[cell(i,"country")]== None:
            total_daily_vaccination[cell(i,"country")] = daily_add
        else:
            total_daily_vaccination[cell(i,"country")] += daily_add

total_daily_vaccination
```

```python
#Q7: Which country is estimated to have completed the most number of vaccinations across the dates mentioned in the dataset?

country_most_vaccination = None

for key in total_daily_vaccination:
    if total_daily_vaccination[key] != None:
        if country_most_vaccination == None or total_daily_vaccination[key] > total_daily_vaccination[country_most_vaccination]:
            country_most_vaccination = key
country_most_vaccination
```

```python
#Q8: How many vaccinations are estimated to have been done across the whole world on each date mentioned in the dataset?

world_daily_vaccination={}

for date in dates:
    vaccination_by_country_dict = daily_vaccinations_on(date)
    vaccination_by_country_dict_without_none ={}
    for country in vaccination_by_country_dict:
        if vaccination_by_country_dict[country]!= None:
            vaccination_by_country_dict_without_none[country]= vaccination_by_country_dict[country]
            
    total_world_vaccinnation_on_date= sum(vaccination_by_country_dict_without_none.values())
    world_daily_vaccination[date] = total_world_vaccinnation_on_date

world_daily_vaccination
```

```python
#Q9: For each country in the dataset, when is the last day for which we have data about the number of fully vaccinated people?

last_vaccination_date_by_country={}
for i in range(len(covid_data)):
    country = cell(i, "country")
    fully_vaxed = cell(i, "people_fully_vaccinated")
    if country not in last_vaccination_date_by_country:
        if fully_vaxed == None:
            last_vaccination_date_by_country[country] = None
        else:
            last_vaccination_date_by_country[country] = cell(i, "date")
    else:
        daily_vaccinations = cell(i, "people_fully_vaccinated")
        if daily_vaccinations != None:
            date = cell(i, "date")
            last_vaccination_date_by_country[country] = date

last_vaccination_date_by_country
```

```python
def fully_vaccinated_by(given_date):
    '''return a dictionary mapping each country to the most recent number of people 
    fully vaccinated by the given date; if no data is available, the value is None.'''
    fully_vax_dict = {}
    for i in range(len(covid_data)):
        country = cell(i, "country")
        date = cell(i, "date")
        fully_vax = cell(i, "people_fully_vaccinated")
        if country not in fully_vax_dict:
            fully_vax_dict[country] = fully_vax
        if int(date[3:5]) <= int(given_date[3:5]) and fully_vax != None:
            fully_vax_dict[country] = fully_vax
        #print(country, date[3:5], date, fully_vax_dict[country])
    return fully_vax_dict
```

```python
def total_vaccinations_by(given_date):
    '''return a dictionary mapping each country to the most recent number of total 
     vaccinations by the given date; if no data is available, the value is None.'''
    total_vax_dict = {}
    for i in range(len(covid_data)):
        country = cell(i, "country")
        date = cell(i, "date")
        total_vax = cell(i, "total_vaccinations")
        if country not in total_vax_dict:
            total_vax_dict[country] = total_vax
        if int(date[3:5]) <= int(given_date[3:5]) and total_vax != None:
            total_vax_dict[country] = total_vax
        #print(country, date[3:5], date, fully_vax_dict[country])
    return total_vax_dict
```

```python
def people_vaccinated_by(given_date):
    '''return a dictionary mapping each country to the most recent number of total 
     vaccinations by the given date; if no data is available, the value is None.'''
    people_vaxed_dict = {}
    for i in range(len(covid_data)):
        country = cell(i, "country")
        date = cell(i, "date")
        people_vaxed = cell(i, "people_vaccinated")
        if country not in people_vaxed_dict:
            people_vaxed_dict[country] = people_vaxed
        if int(date[3:5]) <= int(given_date[3:5]) and people_vaxed != None:
            people_vaxed_dict[country] = people_vaxed
        #print(country, date[3:5], date, fully_vax_dict[country])
    return people_vaxed_dict
```

```python
vaccination_stats = {}

for date in dates: 
    daily_vax_dict = daily_vaccinations_on(date) # a dict mapping each country to number of daily_vaccinations on date
    total_vax_dict = total_vaccinations_by(date) # a dict mapping each country to the most recent number of total_vaccinations on date
    people_vax_dict = people_vaccinated_by(date) # a dict mapping each country to the most recent number of people_vaccinated on date
    fully_vax_dict = fully_vaccinated_by(date) # a dict mapping each country to most recent number of people_fully_vaccinated on date
    for country in countries:
        if country not in vaccination_stats:
            vaccination_stats[country] = {}
        vaccination_stats[country][date] = {}
        vaccination_stats[country][date]['country'] = country
        vaccination_stats[country][date]['date'] = date
        vaccination_stats[country][date]['daily_vaccinations'] = daily_vax_dict[country]
        vaccination_stats[country][date]['total_vaccinations'] = total_vax_dict[country]
        vaccination_stats[country][date]['people_vaccinated'] = people_vax_dict[country]
        vaccination_stats[country][date]['people_fully_vaccinated'] = fully_vax_dict[country]
        vaccination_stats[country][date]['population'] = populations[country]
        # TODO: fill in the rest of the dict to match the example above
        
        
```

```python
#Q17: What is the estimated ratio of people who have been partly vaccinated 
#to people who have been fully vaccinated in each country, as of '10/16/2021'?


ratios_by_country_dict={}
date = '10/16/2021'

for country in countries:
    people_fully_vaccinated_by_date= vaccination_stats[country][date]["people_fully_vaccinated"]
    people_vaccinated_by_date= vaccination_stats[country][date]["people_vaccinated"]
    if people_fully_vaccinated_by_date != None and people_vaccinated_by_date != None:
        people_partially_vaxed_by_date = people_vaccinated_by_date - people_fully_vaccinated_by_date 
        ratio = people_partially_vaxed_by_date/ people_fully_vaccinated_by_date 
        ratios_by_country_dict[country] = ratio
    else: 
        ratios_by_country_dict[country] = None
ratios_by_country_dict

```

```python
#Q18: Which country has the lowest estimated ratio of people 
#who are partly vaccinated to people who are fully vaccinated, as of '10/16/2021'?
best_country = None
lowest_ratio = None


for country in countries:
    ratio = ratios_by_country_dict[country]
    if ratio != None:
        if lowest_ratio == None or lowest_ratio > ratio:
            best_country = country
            lowest_ratio = ratio
best_country
```

```python
def ratio_by_country_per_population (numerator, date):
    ratios_by_country_dict={}
    
    for country in countries:
        numerator_by_date= vaccination_stats[country][date][numerator]
        population= populations[country]
        if numerator_by_date != None and population != None:
            ratio = numerator_by_date/ population
            ratios_by_country_dict[country] = ratio
        else: 
            ratios_by_country_dict[country] = None
    return ratios_by_country_dict

total_vax_ratio_per_population_by_country = ratio_by_country_per_population ("total_vaccinations", '10/14/2021')

```

```python
def country_with_highest_standard_given_dict (dictionary):

    best_country = None
    highest_standard = None

    for country in countries:
        standard = dictionary[country]
        if standard != None:
            if highest_standard  == None or highest_standard  < standard:
                best_country = country
                highest_standard  = standard
    return best_country
```
